#1. Análise do faturamento


```python
faturmaento_loja1= loja1['Preço'].sum()
print(f"O faturamento completo da Loja 1 foi: R$ {faturmaento_loja1:.2f}")
```

```python
# prompt: faturamento completo da loja2

faturamento_loja2 = loja2['Preço'].sum()
print(f"O faturamento completo da Loja 2 foi: R$ {faturamento_loja2:.2f}")
```

```python
faturamento_loja3 = loja3['Preço'].sum()
print(f"O faturamento completo da Loja 3 foi: R$ {faturamento_loja3:.2f}")
```

```python
faturamento_loja4 = loja4['Preço'].sum()
print(f"O faturamento completo da Loja 4 foi: R$ {faturamento_loja4:.2f}")
```

```python
# prompt: comparar o faturamento das 4 loja

print("\nComparação do Faturamento das Lojas:")
print(f"Loja 1: R$ {faturmaento_loja1:.2f}")
print(f"Loja 2: R$ {faturamento_loja2:.2f}")
print(f"Loja 3: R$ {faturamento_loja3:.2f}")
print(f"Loja 4: R$ {faturamento_loja4:.2f}")

faturamentos = {
    "Loja 1": faturmaento_loja1,
    "Loja 2": faturamento_loja2,
    "Loja 3": faturamento_loja3,
    "Loja 4": faturamento_loja4
}

loja_maior_faturamento = max(faturamentos, key=faturamentos.get)
maior_valor_faturamento = faturamentos[loja_maior_faturamento]

loja_menor_faturamento = min(faturamentos, key=faturamentos.get)
menor_valor_faturamento = faturamentos[loja_menor_faturamento]

print(f"\nA loja com maior faturamento é a {loja_maior_faturamento} com R${maior_valor_faturamento:.2f}")
print(f"\nA loja com menor faturamento é a {loja_menor_faturamento} com R${menor_valor_faturamento:.2f}")
```

```python
# Dados de faturamento das lojas
lojas = list(faturamentos.keys())
valores_faturamento = list(faturamentos.values())
# Criar o gráfico de barras verticais
plt.figure(figsize=(8, 6))
plt.bar(lojas, valores_faturamento, color='skyblue')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.title('Comparação de Faturamento das Lojas')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Grafico de pizza  para venda por categoria

vendas_categoria_l4 = loja4.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)

# Combine os dados de todas as lojas
vendas_por_categoria_total = pd.concat([
    loja1.groupby('Categoria do Produto')['Preço'].sum(),
    loja2.groupby('Categoria do Produto')['Preço'].sum(),
    loja3.groupby('Categoria do Produto')['Preço'].sum(),
    loja4.groupby('Categoria do Produto')['Preço'].sum()
]).groupby(level=0).sum() # Agrupe por categoria e some os valores

# Gere o gráfico de pizza
plt.figure(figsize=(10, 8))
plt.pie(vendas_por_categoria_total, labels=vendas_por_categoria_total.index, autopct='%1.1f%%', startangle=140)
plt.title('Vendas por Categoria (Total das Lojas)')
plt.axis('equal')  # Garante que o gráfico de pizza seja um círculo.
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
avaliacao_loja1 = loja1['Avaliação da compra'].mean()
avaliacao_loja2 = loja2['Avaliação da compra'].mean()
avaliacao_loja3 = loja3['Avaliação da compra'].mean()
avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print("\n=== MÉDIA DE AVALIAÇÃO POR LOJA ===")
print(f"Loja 1: {avaliacao_loja1:.2f}")
print(f"Loja 2: {avaliacao_loja2:.2f}")
print(f"Loja 3: {avaliacao_loja3:.2f}")
print(f"Loja 4: {avaliacao_loja4:.2f}")
```

```python
# prompt: grafico de barra sobre a media de cada loja

medias_avaliacao = [avaliacao_loja1, avaliacao_loja2, avaliacao_loja3, avaliacao_loja4]

plt.figure(figsize=(8, 4))
plt.bar(lojas, medias_avaliacao, color= 'skyblue')
plt.xlabel('Lojas')
plt.ylabel('Média de Avaliação')
plt.title('Média de Avaliação por Loja')
plt.ylim(0, 5) # Definir o limite do eixo y de 0 a 5 (escala de avaliação)
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
#Reunir todas as lojas
lojas = pd.concat([loja1, loja2, loja3, loja4])

#contagem da compra dos produtos
contagem_produtos = lojas['Produto'].value_counts()

#produtos mais vendidos
produtos_mais_vendidos = contagem_produtos.head(5)
print(produtos_mais_vendidos)

```

```python
#produtos menos vendidos
produtos_menos_vendidos = contagem_produtos.tail(5)
print(produtos_menos_vendidos)
```

# 5. Frete Médio por Loja


```python
#Frete Médio por Loja- Qual Loja tem o menor custo de frete?
frete_loja1 = loja1['Frete'].mean()
frete_loja2 = loja2['Frete'].mean()
frete_loja3 = loja3['Frete'].mean()
frete_loja4 = loja4['Frete'].mean()

print ("\n=== FRETE MÉDIO POR LOJA ===")
print(f"Loja 1: R$ {frete_loja1:.2f}")
print(f"Loja 2: R$ {frete_loja2:.2f}")
print(f"Loja 3: R$ {frete_loja3:.2f}")
print(f"Loja 4: R$ {frete_loja4:.2f}")
```

```python
#Criando o gráfico de comparação
lojas =['Loja1', 'Loja2', 'Loja3', 'Loja4']
frete_medio = [frete_loja1, frete_loja2, frete_loja3, frete_loja4]

plt.figure(figsize = (8,6))
plt.bar(lojas, frete_medio, color=['steelblue', 'skyblue', 'lightskyblue','cadetblue'])
plt.xlabel('Lojas')
plt.ylabel('Frete (R$)')
plt.title('Frete Médio das Lojas')
plt.show()
```
